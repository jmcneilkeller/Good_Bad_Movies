```python
import implicit
import pandas as pd
import scipy
```

```python
df = pd.read_csv('item_item_final.csv',index_col=0)
df.head(30)
```

```python
sparse = scipy.sparse.csr_matrix(df.values)
```

```python
als_model = implicit.als.AlternatingLeastSquares()
```

```python
als_model.fit(sparse)
```

```python
sim_items = als_model.similar_items(66)
```
